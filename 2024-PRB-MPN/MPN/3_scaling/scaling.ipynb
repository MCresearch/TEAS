```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda_ = np.array([0.25, 0.5, 1, 2, 3])
energy = np.array([9.507915e-02, 3.803166e-01, 1.521266e+00, 6.085066e+00, 1.369140e+01]) * 13.6056923

def scaling(lambda_, E0):
    return E0 * lambda_ ** 2
```

```python
plt.figure(dpi=200)
plt.plot(np.linspace(0, 4, 100), scaling(np.linspace(0, 4, 100), energy[2]), label=r'$\lambda^2T_{\rm{s}}[\rho]$', color='gray', linestyle='--', zorder=0)
plt.scatter(lambda_, energy, label='MPN KEDF', color='r', marker='*', s=50, zorder=1)
plt.xlim(0,3.5)
plt.legend(fontsize=12)
plt.xlabel('$\lambda$', fontsize=14)
plt.ylabel(r'$T_{\rm{s}}[\rho_{\lambda}]$ (eV)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
