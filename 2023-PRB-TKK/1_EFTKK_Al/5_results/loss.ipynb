```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pylab
```

```python
def readLoss(file):
    data = np.loadtxt(file, skiprows=3)
    loss_ = data[:,-2]
    miniloss_ = data[:,-1]
    return np.array([loss_, miniloss_])
```

```python
files = ['./1log', './3log', './4log']
lables = [8, 12, 16]
```

```python
# files = ['0_gpq_gpnl', '1_gpq_gnl', '2_gpq_gqnl', '3_all', '4_pq_gnl', '5_ggnl']
# labels = ["$\gamma, p, q, \gamma_{nl}, p_{nl}$", "$\gamma, p, q, \gamma_{nl}$", "$\gamma, p, q, \gamma_{nl}, q_{nl}$", "$\gamma, p, q, \gamma_{nl}, p_{nl}, q_{nl}$", "$p, q, \gamma_{nl}$", "$\gamma, \gamma_{nl}$"]
colors = ['red', 'skyblue', 'seagreen', 'm', 'royalblue', 'orange']

plt.figure(dpi=250)
cm = pylab.get_cmap('gist_rainbow')
Nnet = len(files)

# for i in [0,1,2,3,4]:
for i in range(len(files)):
    loss = readLoss(files[i])[1]

    color = cm(i/Nnet)
    plt.plot(np.arange(len(loss)), loss, linestyle='--', color=color, label="$\lambda_c=$"+str(lables[i]))
    # plt.plot(np.arange(len(data[1])), data[1], linestyle='--', color=colors[i], label=title)
    # plt.plot(data[0], data[2], linestyle=':',  color=colors[i], label=labels[i] + " (vali)")
plt.grid(linestyle=':')
plt.yscale("log")
# plt.xscale("log")
# plt.xlim(499900,500000)
# plt.xlim(60000*8-100,60000*8)
plt.xlim(0,21000)
plt.legend(fontsize=14)
plt.xlabel("Steps", fontsize=14)
plt.ylabel("Residual", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title(r"$\rm{TKK^m}$", fontsize=14)
plt.show()
```
