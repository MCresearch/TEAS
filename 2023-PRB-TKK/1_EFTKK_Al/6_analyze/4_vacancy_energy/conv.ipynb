```python
import numpy as np
import matplotlib.pyplot as plt
import xlrd
```

```python
data = xlrd.open_workbook('./vacancy_energy_fcc.xls')
sheet = data.sheet_by_index(0)
markers = ['D', 'o', 'd', 's', 'p', 'h']
colors = ['k', 'orchid', 'darkseagreen', 'r', 'royalblue', 'sandybrown']

WT = np.array([
1.2369310573499774,
1.3469077056624883,
1.4072071865624594,
1.4468727007501911,
1.4726021564820257,
1.479163177891678,
1.480490227455448,
1.4813083670160268,
1.4818284037173726
])

WGC = np.array([
0.7063852133750288,
0.7404439448875451,
0.8087243270188083,
0.8736254881875993,
0.881456087389779,
0.8827809115246055,
0.8821158670180012,
0.8821269293184741,
0.8823160632600775
])

KS = np.array([
0.796,
0.757,
0.747,
0.794,
0.807,
0.808
])
```

```python
natom = np.array([1,2,4,8,27,64,125,6**3,7**3]) * 4 - 1
lambdac = np.array([8,12,16,20])
plt.figure(dpi=200)
plt.plot(natom[:6], KS, '--', color=colors[0], marker=markers[0], label='KS-BLPS', markerfacecolor='white')
plt.plot(natom, WGC, '--', color=colors[1], marker=markers[1], label='WGC', markerfacecolor='white')
plt.plot(natom, WT, '--', color=colors[2], marker=markers[2], label='WT', markerfacecolor='white')
for i in range(2,-1,-1):
    # ev = np.array([float(sheet.col_values)])
    plt.plot(natom, np.array(sheet.row_values(i+1)[1:], dtype=np.float64), '--', color=colors[5-i], marker=markers[5-i], label=r'$\rm{TKK}^{\rm{m}}_{\rm{%d}}$' % lambdac[i], markerfacecolor='white')
plt.legend(ncol=2,fontsize=12,loc=10)
plt.xscale('log')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Number of atoms",fontsize=16)
plt.ylabel("Vacancy energy (eV)",fontsize=16)
# plt.title('Vacancy energy of fcc Al')
plt.ylim(0.5, 1.7)
plt.text(50, 1.6, 'fcc Al',fontsize=14)
plt.text(3.5, 1.6, '(a)', fontsize=14)
# plt.show()
plt.savefig('conv.png', dpi=350)
```
