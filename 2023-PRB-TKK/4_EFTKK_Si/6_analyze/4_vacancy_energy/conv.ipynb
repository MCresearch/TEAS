```python
import numpy as np
import matplotlib.pyplot as plt
import xlrd
```

```python
data = xlrd.open_workbook('./vacancy_energy_CD.xls')
sheet = data.sheet_by_index(0)
markers = ['D', 'o', 'd', 's', 'p', 'h']
colors = ['k', 'orchid', 'darkseagreen', 'r', 'royalblue', 'sandybrown']

WT = np.array([
-0.5715835020375835,
-0.4533815587499248,
-0.3462578347498493,
-0.3207795058369811,
-0.3151870860019699,
-0.312669356004335,
-0.31173965576454066,
-0.31121235562022775
])

KS = np.array([
2.735,
3.026,
3.240,
3.517,
3.417
])

HC = np.array([
2.63016844,
2.277395705,
# 1.452641826,
# 0.989259599,
# -5.201419857,
# -6.375796814,
# 4.818054478,
# 5.155106538
])
```

```python
natom = np.array([1,2,8,27,64,125,6**3,7**3]) * 8 - 1
lambdac = np.array([8,12,16,20])
plt.figure(dpi=200)
plt.plot(natom[:5], KS, '--', color=colors[0], marker=markers[0], label='KS-BLPS', markerfacecolor='white')
plt.plot(natom[:2], HC, '--', color=colors[1], marker=markers[1], label='HC', markerfacecolor='white')
plt.plot(natom, WT, '--', color=colors[2], marker=markers[2], label='WT', markerfacecolor='white')
# plt.plot(natom, WGC, '.--', label='WGC')
for i in range(2,-1,-1):
    # ev = np.array([float(sheet.col_values)])
    plt.plot(natom, np.array(sheet.row_values(i+1)[1:], dtype=np.float64), '--', color=colors[5-i], marker=markers[5-i], label=r'$\rm{TKK}^{\rm{s}}_{\rm{%d}}$' % lambdac[i], markerfacecolor='white')
    # plt.plot(natom, np.array(sheet.row_values(i+1)[1:], dtype=np.float64), '.--', label=r'$\lambda_{c} = %d$' % lambdac[i])
plt.legend(ncol=2,fontsize=12,loc=10)
plt.xscale('log')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Number of atoms",fontsize=16)
plt.ylabel("Vacancy energy (eV)",fontsize=16)
# plt.title('Vacancy energy of CD Si')
plt.ylim(-1.5, 4.5)
plt.text(100, 4, 'CD Si', fontsize=14)
plt.text(7.2, 4, '(b)', fontsize=14)
# plt.show()
plt.savefig('conv.png', dpi=350)

```
